```python
import pandas as pd
from geopy.distance import geodesic
```

```python
businesses = pd.read_csv("yelp_academic_dataset_business.csv")
```

```python
businesses = businesses[(businesses['city'] == 'Philadelphia') & (businesses['categories'].str.contains('Restaurants', na=False))]
businesses
```

```python
# straight line distance in km
def geopy_distance(lat1, lon1, restaurant):

    coords_1 = (lat1, lon1)
    coords_2 = (float(businesses[businesses['business_id'] == restaurant]['latitude']), float(businesses[businesses['business_id'] == restaurant]['longitude']))
    
    return geodesic(coords_1, coords_2).kilometers
```

```python
geopy_distance(40, -80, 'WnT9NIzQgLlILjPT0kEcsQ')
```

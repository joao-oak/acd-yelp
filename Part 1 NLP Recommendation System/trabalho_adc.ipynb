```python
ratings = ratings.reset_index(drop=True)

# Passo 3: Agrupar por restaurante e somar os vetores TF-IDF
def sum_tfidf_vectors_for_restaurant(group):
    # Obter os índices das reviews do restaurante
    indices = group.index
    # Somar os vetores TF-IDF correspondentes a essas reviews
    return tfidf_matrix[indices].sum(axis=0)

tfidf_by_restaurant = ratings.groupby('business_id').apply(sum_tfidf_vectors_for_restaurant).reset_index()

# Exibir o resultado
print(tfidf_by_restaurant.head())
```
